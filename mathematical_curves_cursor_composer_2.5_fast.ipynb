# 数学曲线大全

本 Notebook 汇总常见数学曲线的定义、参数方程与可视化，涵盖：

- 笛卡尔坐标系下的基本曲线
- 参数曲线
- 极坐标曲线
- 经典特殊曲线
- 三维空间曲线

依赖：`numpy`、`matplotlib`

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['figure.dpi'] = 100


def plot_curve(x, y, title, xlabel='x', ylabel='y', **kwargs):
    """绘制二维曲线。"""
    fig, ax = plt.subplots()
    ax.plot(x, y, **kwargs)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.grid(True, alpha=0.3)
    ax.set_aspect('equal', adjustable='box')
    plt.tight_layout()
    plt.show()


def plot_polar(r, theta, title, **kwargs):
    """绘制极坐标曲线。"""
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
    ax.plot(theta, r, **kwargs)
    ax.set_title(title, pad=20)
    plt.tight_layout()
    plt.show()


def plot_parametric(x, y, title, **kwargs):
    plot_curve(x, y, title, **kwargs)


def plot_3d(x, y, z, title):
    """绘制三维曲线。"""
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x, y, z)
    ax.set_title(title)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    plt.tight_layout()
    plt.show()
```

## 1. 笛卡尔坐标系 — 圆锥曲线与基本函数

```python
# 1.1 直线 y = kx + b
x = np.linspace(-5, 5, 200)
plot_curve(x, 2 * x + 1, 'y = 2x + 1') #直线: 
```

```python
# 1.2 圆: x² + y² = r²
t = np.linspace(0, 2 * np.pi, 400)
r = 3
plot_parametric(r * np.cos(t), r * np.sin(t), f'x² + y² = {r**2}')  #圆: 
```

```python
# 1.3 椭圆: x²/a² + y²/b² = 1
a, b = 5, 3
plot_parametric(a * np.cos(t), b * np.sin(t), f'x²/{a**2} + y²/{b**2} = 1') #椭圆: 
```

```python
# 1.4 抛物线: y = ax²
x = np.linspace(-4, 4, 300)
plot_curve(x, 0.5 * x**2, 'y = 0.5x²') #抛物线: 
```

```python
# 1.5 双曲线: x²/a² - y²/b² = 1
a, b = 3, 2
x_pos = np.linspace(a, 8, 200)
x_neg = np.linspace(-8, -a, 200)
y_pos = b * np.sqrt(x_pos**2 / a**2 - 1)
y_neg = -y_pos
fig, ax = plt.subplots()
ax.plot(x_pos, y_pos, 'b')
ax.plot(x_pos, y_neg, 'b')
ax.plot(x_neg, y_pos, 'b')
ax.plot(x_neg, y_neg, 'b')
ax.set_title(f'x²/{a**2} - y²/{b**2} = 1') #双曲线: 
ax.grid(True, alpha=0.3)
ax.set_aspect('equal', adjustable='box')
plt.tight_layout()
plt.show()
```

```python
# 1.6 正弦曲线
x = np.linspace(0, 4 * np.pi, 500)
plot_curve(x, np.sin(x), 'y = sin(x)') #正弦曲线: 
```

```python
# 1.7 余弦曲线
plot_curve(x, np.cos(x), 'y = cos(x)') #余弦曲线: 
```

```python
# 1.8 正切曲线
x = np.linspace(-1.5 * np.pi, 1.5 * np.pi, 1000)
y = np.tan(x)
y[np.abs(y) > 10] = np.nan
plot_curve(x, y, 'y = tan(x)') #正切曲线: 
```

```python
# 1.9 指数函数
x = np.linspace(-3, 3, 300)
plot_curve(x, np.exp(x), 'y = eˣ') #指数函数: 
```

```python
# 1.10 对数函数
x = np.linspace(0.01, 10, 300)
plot_curve(x, np.log(x), 'y = ln(x)') #对数函数: 
```

```python
# 1.11 高斯函数（正态分布）
x = np.linspace(-5, 5, 400)
mu, sigma = 0, 1
plot_curve(x, np.exp(-(x - mu)**2 / (2 * sigma**2)) / (sigma * np.sqrt(2 * np.pi)),
           '') #高斯函数（标准正态分布）
```

```python
# 1.12 双曲正弦 / 双曲余弦
x = np.linspace(-3, 3, 300)
fig, ax = plt.subplots()
ax.plot(x, np.sinh(x), label='sinh(x)')
ax.plot(x, np.cosh(x), label='cosh(x)')
ax.set_title('') #双曲函数
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 1.13 悬链线: y = a·cosh(x/a)
a = 2
x = np.linspace(-4, 4, 300)
plot_curve(x, a * np.cosh(x / a), f'y = {a}·cosh(x/{a})') #悬链线: 
```

```python
# 1.14 符号函数与绝对值
x = np.linspace(-5, 5, 500)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(x, np.sign(x))
axes[0].set_title('sign(x)') #符号函数 
axes[0].grid(True, alpha=0.3)
axes[1].plot(x, np.abs(x))
axes[1].set_title('|x|') #绝对值函数 
axes[1].grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 2. 参数曲线

```python
# 2.1 摆线 (Cycloid): x = r(t - sin t), y = r(1 - cos t)
t = np.linspace(0, 4 * np.pi, 500)
r = 2
plot_parametric(r * (t - np.sin(t)), r * (1 - np.cos(t)), '(Cycloid)') #摆线 
```

```python
# 2.2 心形线 (Cardioid): x = a(1-cos t)cos t, y = a(1-cos t)sin t
a = 2
plot_parametric(a * (1 - np.cos(t)) * np.cos(t),
                a * (1 - np.cos(t)) * np.sin(t),
                '(Cardioid)') #心形线 
```

```python
# 2.3 星形线 (Astroid): x = a cos³t, y = a sin³t
a = 4
plot_parametric(a * np.cos(t)**3, a * np.sin(t)**3, '(Astroid)') #星形线 
```

```python
# 2.4 外摆线 (Epicycloid)
R, r, d = 5, 2, 3
k = (R + r) / r
x = (R + r) * np.cos(t) - d * np.cos(k * t)
y = (R + r) * np.sin(t) - d * np.sin(k * t)
plot_parametric(x, y, '(Epicycloid)') #外摆线 
```

```python
# 2.5 内摆线 (Hypocycloid)
R, r, d = 5, 1.5, 4
k = (R - r) / r
x = (R - r) * np.cos(t) + d * np.cos(k * t)
y = (R - r) * np.sin(t) - d * np.sin(k * t)
plot_parametric(x, y, '(Hypocycloid)') #内摆线 
```

```python
# 2.6 李萨如曲线 (Lissajous): x = sin(at + δ), y = sin(bt)
t = np.linspace(0, 2 * np.pi, 1000)
a, b, delta = 3, 4, np.pi / 2
plot_parametric(np.sin(a * t + delta), np.sin(b * t),
                f'(a={a}, b={b}, δ=π/2)') #李萨如曲线 
```

```python
# 2.7 螺旋线 (Archimedean Spiral): r = a·θ
theta = np.linspace(0, 6 * np.pi, 800)
a = 0.3
plot_parametric(a * theta * np.cos(theta), a * theta * np.sin(theta),
                '(Archimedean Spiral)') #阿基米德螺线 
```

```python
# 2.8 对数螺线 (Logarithmic Spiral): r = a·e^(bθ)
theta = np.linspace(0, 4 * np.pi, 800)
a, b = 0.5, 0.15
r = a * np.exp(b * theta)
plot_parametric(r * np.cos(theta), r * np.sin(theta), '(Logarithmic Spiral)') #对数螺线 
```

```python
# 2.9 费马螺线 (Fermat's Spiral): r² = a²θ
theta = np.linspace(0, 10 * np.pi, 1000)
a = 0.8
r = a * np.sqrt(theta)
plot_parametric(r * np.cos(theta), r * np.sin(theta), "(Fermat's Spiral)") #费马螺线 
```

```python
# 2.10 蝴蝶曲线 (Butterfly Curve)
t = np.linspace(0, 12 * np.pi, 5000)
r = np.exp(np.cos(t)) - 2 * np.cos(4 * t) + np.sin(t / 12)**5
plot_parametric(r * np.cos(t), r * np.sin(t), '(Butterfly Curve)') #蝴蝶曲线 
```

```python
# 2.11 蔓叶线 (Folium of Descartes): x³ + y³ = 3axy
t = np.linspace(-5, 5, 2000)
a = 1
x = 3 * a * t / (1 + t**3)
y = 3 * a * t**2 / (1 + t**3)
plot_parametric(x, y, '(Folium of Descartes)') #蔓叶线 
```

```python
# 2.12 箕舌线 (Witch of Agnesi): y = 8a³ / (x² + 4a²)
a = 1
x = np.linspace(-10, 10, 500)
plot_curve(x, 8 * a**3 / (x**2 + 4 * a**2), '(Witch of Agnesi)') #箕舌线 
```

```python
# 2.13 悬索线参数形式 / 曳物线 (Tractrix)
t = np.linspace(-3, 3, 400)
a = 2
x = a * (t - np.tanh(t))
y = a / np.cosh(t)
plot_parametric(x, y, '(Tractrix)') #曳物线 
```

```python
# 2.14 圆渐开线 (Involute of Circle)
t = np.linspace(0, 4 * np.pi, 500)
a = 2
plot_parametric(a * (np.cos(t) + t * np.sin(t)),
                a * (np.sin(t) - t * np.cos(t)),
                '(Involute of Circle)') #圆渐开线 
```

## 3. 极坐标曲线

```python
# 3.1 玫瑰线 (Rose Curve): r = a·cos(nθ) 或 r = a·sin(nθ)
theta = np.linspace(0, 2 * np.pi, 1000)
n = 5
plot_polar(np.cos(n * theta), theta, f'r = cos({n}θ)') #玫瑰线: 
```

```python
# 3.2 极坐标心形线: r = a(1 - cos θ)
a = 2
plot_polar(a * (1 - np.cos(theta)), theta, 'r = 2(1 - cos θ)') #极坐标心形线: 
```

```python
# 3.3 双纽线 (Lemniscate of Bernoulli): r² = a² cos(2θ)
a = 3
r = a * np.sqrt(np.abs(np.cos(2 * theta)))
#r[np.isnan(r)] = 0
plot_polar(r, theta, '(Lemniscate of Bernoulli)') #双纽线 
```

```python
# 3.4 极坐标阿基米德螺线: r = a·θ
theta = np.linspace(0, 8 * np.pi, 1000)
plot_polar(0.5 * theta, theta, 'r = 0.5θ') #极坐标阿基米德螺线: 
```

```python
# 3.5 极坐标对数螺线: r = a·e^(bθ)
theta = np.linspace(0, 3 * np.pi, 800)
plot_polar(0.5 * np.exp(0.2 * theta), theta, 'r = 0.5·e^(0.2θ)') #极坐标对数螺线: 
```

```python
# 3.6 四叶玫瑰线
plot_polar(np.sin(2 * theta), theta, 'r = sin(2θ)') #四叶玫瑰线: 
```

```python
# 3.7 笛卡尔卵形线 (Cartesian Oval) — 极坐标近似
a, b, c = 2, 3, 1
r = (a * b + c**2 * np.cos(2 * theta)) / (a + c * np.cos(theta))
plot_polar(r, theta, '(Cartesian Oval)') #笛卡尔卵形线 
```

```python
# 3.8 圆锥螺线 (Conchoid of Nicomedes) — 极坐标形式
a, b = 1, 2
r = a + b / np.cos(theta)
r[np.abs(r) > 20] = np.nan
plot_polar(r, theta, '(Conchoid of Nicomedes)') #尼科米德斯蚌线 
```

## 4. 经典特殊曲线

```python
# 4.1 笛卡尔叶形线 (Folium) — 隐式形式可视化
x = np.linspace(-3, 3, 300)
y = np.linspace(-3, 3, 300)
X, Y = np.meshgrid(x, y)
a = 1
Z = X**3 + Y**3 - 3 * a * X * Y
fig, ax = plt.subplots()
ax.contour(X, Y, Z, levels=[0], colors='blue')
ax.set_title('x³ + y³ = 3axy') #笛卡尔叶形线: 
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 4.2 卡西尼卵形线 (Cassini Oval): (x²+y²)² - 2c²(x²-y²) = a⁴ - c⁴
x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)
a, c = 1.5, 1
Z = (X**2 + Y**2)**2 - 2 * c**2 * (X**2 - Y**2) - (a**4 - c**4)
fig, ax = plt.subplots()
ax.contour(X, Y, Z, levels=[0], colors='purple')
ax.set_title('(Cassini Oval)')  #卡西尼卵形线 
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 4.3 椭圆积分相关 — 椭圆: 隐式 x²/a² + y²/b² = 1 (多椭圆族)
fig, ax = plt.subplots()
for b in [1, 2, 3, 4]:
    ax.plot(5 * np.cos(t), b * np.sin(t), label=f'b={b}')
ax.set_title('x²/25 + y²/b² = 1') #椭圆族: 
ax.legend()
ax.grid(True, alpha=0.3)
ax.set_aspect('equal')
plt.tight_layout()
plt.show()
```

```python
# 4.4 超椭圆 (Superellipse): |x/a|^n + |y/b|^n = 1
n = 2.5
a, b = 4, 3
t = np.linspace(0, 2 * np.pi, 500)
x = a * np.sign(np.cos(t)) * np.abs(np.cos(t))**(2 / n)
y = b * np.sign(np.sin(t)) * np.abs(np.sin(t))**(2 / n)
plot_parametric(x, y, f'(n={n})') #超椭圆 
```

```python
# 4.5 尖点线 (Cusp): y² = x³
x = np.linspace(0, 5, 300)
y_pos = np.sqrt(x**3)
fig, ax = plt.subplots()
ax.plot(x, y_pos, 'r')
ax.plot(x, -y_pos, 'r')
ax.set_title('(Semicubical Parabola): y² = x³') #尖点线 
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 4.6 环索线 (Limaçon): r = b - a·cos θ
a, b = 2, 3
plot_polar(b - a * np.cos(theta), theta, '(Limaçon): r = 3 - 2cos θ') #环索线 
```

```python
# 4.7 三次抛物线
x = np.linspace(-3, 3, 300)
plot_curve(x, x**3 / 3, 'y = x³/3') #三次抛物线: 
```

```python
# 4.8 反比例函数
x = np.linspace(-5, -0.1, 200)
x2 = np.linspace(0.1, 5, 200)
fig, ax = plt.subplots()
ax.plot(x, 1 / x, 'b')
ax.plot(x2, 1 / x2, 'b')
ax.set_title('y = 1/x') #反比例函数: 
ax.set_ylim(-10, 10)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 4.9 锯齿波 / 方波 / 三角波
x = np.linspace(0, 4 * np.pi, 500)
fig, axes = plt.subplots(3, 1, figsize=(10, 8))
axes[0].plot(x, 2 * (x / (2 * np.pi) - np.floor(0.5 + x / (2 * np.pi))))
axes[0].set_title('') #锯齿波
axes[0].grid(True, alpha=0.3)
axes[1].plot(x, np.sign(np.sin(x)))
axes[1].set_title('') #方波
axes[1].grid(True, alpha=0.3)
axes[2].plot(x, 2 * np.arcsin(np.sin(x)) / np.pi)
axes[2].set_title('') #三角波
axes[2].grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 4.10 概率密度函数 — 多种分布
x = np.linspace(-1, 10, 500)
fig, ax = plt.subplots()
lam = 2
ax.plot(x, lam * np.exp(-lam * x), label=f'λ={lam}') #指数分布 
x2 = np.linspace(0, 10, 500)
k, theta = 2, 2
import math
ax.plot(x2, x2**(k - 1) * np.exp(-x2 / theta) / (math.gamma(k) * theta**k),
        label=f'Gamma(k={k}, θ={theta})')
ax.set_title('') #概率密度曲线
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 5. 三维空间曲线

```python
# 5.1 圆柱螺旋线 (Helix): x = cos t, y = sin t, z = t
t = np.linspace(0, 4 * np.pi, 500)
plot_3d(np.cos(t), np.sin(t), t, '(Helix)') #圆柱螺旋线 
```

```python
# 5.2 圆锥螺旋线
t = np.linspace(0, 6 * np.pi, 500)
r = t / (2 * np.pi)
plot_3d(r * np.cos(t), r * np.sin(t), t / 3, '') #圆锥螺旋线
```

```python
# 5.3 环面纽结 (Torus Knot, p=2, q=3)
t = np.linspace(0, 2 * np.pi, 1000)
p, q, R, r = 2, 3, 3, 1
x = (R + r * np.cos(q * t)) * np.cos(p * t)
y = (R + r * np.cos(q * t)) * np.sin(p * t)
z = r * np.sin(q * t)
plot_3d(x, y, z, '(Torus Knot, p=2, q=3)') #环面纽结 
```

```python
# 5.4 Viviani 曲线
t = np.linspace(0, 4 * np.pi, 500)
R = 3
x = R * (1 + np.cos(t))
y = R * np.sin(t)
z = 2 * R * np.sin(t / 2)
plot_3d(x, y, z, 'Viviani') # 曲线
```

```python
# 5.5 空间椭圆
t = np.linspace(0, 2 * np.pi, 500)
plot_3d(3 * np.cos(t), 2 * np.sin(t), 0.5 * np.sin(2 * t), '') #空间椭圆曲线
```

## 6. 曲线总览图

```python
# 多曲线对比总览
fig, axes = plt.subplots(3, 4, figsize=(16, 12))
curves = [
    ('', np.cos(t), np.sin(t)), #圆
    ('', 2*np.cos(t), np.sin(t)), #椭圆
    ('', (1-np.cos(t))*np.cos(t), (1-np.cos(t))*np.sin(t)), #心形线
    ('', np.cos(t)**3, np.sin(t)**3),  #星形线
    ('', np.sin(3*t+np.pi/2), np.sin(4*t)), #李萨如
    ('', t-np.sin(t), 1-np.cos(t)), #摆线
    ('', 0.2*t*np.cos(t), 0.2*t*np.sin(t)), #阿基米德螺线
    ('', np.sqrt(np.abs(np.cos(2*t)))*np.cos(t), #双纽线
     np.sqrt(np.abs(np.cos(2*t)))*np.sin(t)),
    ('', np.cos(3*t)*np.cos(t), np.cos(3*t)*np.sin(t)), #玫瑰线
    ('', t, np.cosh(t)), #悬链线
    ('', t/2, np.sin(t)), #正弦
    ('', t, 0.1*t**2), #抛物线
]
for ax, (name, cx, cy) in zip(axes.flat, curves):
    ax.plot(cx, cy, linewidth=1.5)
    ax.set_title(name, fontsize=10)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.2)
    ax.tick_params(labelsize=7)
plt.suptitle('', fontsize=14, y=1.01) #数学曲线总览
plt.tight_layout()
plt.show()
```
