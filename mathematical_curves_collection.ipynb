# Mathematical Curves Collection
## 已知数学曲线集合

This notebook demonstrates various well-known mathematical curves and their visualizations using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches
import seaborn as sns

# Set up the plotting style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')
```

## 1. Circle (圆)
**Parametric equations:**
- x = r * cos(t)
- y = r * sin(t)

```python
fig, ax = plt.subplots(figsize=(6, 6))
t = np.linspace(0, 2*np.pi, 1000)
r = 1
x = r * np.cos(t)
y = r * np.sin(t)
ax.plot(x, y, 'b-', linewidth=2, label='Circle')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Circle (圆)')
plt.show()
```

## 2. Ellipse (椭圆)
**Parametric equations:**
- x = a * cos(t)
- y = b * sin(t)

```python
fig, ax = plt.subplots(figsize=(8, 6))
t = np.linspace(0, 2*np.pi, 1000)
a, b = 2, 1
x = a * np.cos(t)
y = b * np.sin(t)
ax.plot(x, y, 'r-', linewidth=2, label='Ellipse (a=2, b=1)')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Ellipse (椭圆)')
plt.show()
```

## 3. Spiral (螺旋线)
### 3.1 Archimedes Spiral (阿基米德螺旋)
**Parametric equations:**
- x = a * t * cos(t)
- y = a * t * sin(t)

```python
fig, ax = plt.subplots(figsize=(8, 8))
t = np.linspace(0, 4*np.pi, 1000)
a = 0.5
x = a * t * np.cos(t)
y = a * t * np.sin(t)
ax.plot(x, y, 'g-', linewidth=2, label='Archimedes Spiral')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Archimedes Spiral (阿基米德螺旋)')
plt.show()
```

### 3.2 Logarithmic Spiral (对数螺旋)
**Parametric equations:**
- x = a * exp(b*t) * cos(t)
- y = a * exp(b*t) * sin(t)

```python
fig, ax = plt.subplots(figsize=(8, 8))
t = np.linspace(0, 4*np.pi, 1000)
a, b = 0.1, 0.1
x = a * np.exp(b*t) * np.cos(t)
y = a * np.exp(b*t) * np.sin(t)
ax.plot(x, y, 'm-', linewidth=2, label='Logarithmic Spiral')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Logarithmic Spiral (对数螺旋)')
plt.show()
```

## 4. Cardioid (心形线)
**Parametric equations:**
- x = a * (2*cos(t) - cos(2*t))
- y = a * (2*sin(t) - sin(2*t))

```python
fig, ax = plt.subplots(figsize=(8, 8))
t = np.linspace(0, 2*np.pi, 1000)
a = 0.5
x = a * (2*np.cos(t) - np.cos(2*t))
y = a * (2*np.sin(t) - np.sin(2*t))
ax.plot(x, y, 'r-', linewidth=2, label='Cardioid')
ax.fill(x, y, alpha=0.3, color='red')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Cardioid (心形线)')
plt.show()
```

## 5. Astroid (星形线)
**Parametric equations:**
- x = a * cos³(t)
- y = a * sin³(t)

```python
fig, ax = plt.subplots(figsize=(8, 8))
t = np.linspace(0, 2*np.pi, 1000)
a = 1
x = a * np.cos(t)**3
y = a * np.sin(t)**3
ax.plot(x, y, 'c-', linewidth=2, label='Astroid')
ax.fill(x, y, alpha=0.3, color='cyan')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Astroid (星形线)')
plt.show()
```

## 6. Rose Curve / Rhodonea (玫瑰曲线)
**Polar equation:**
- r = a * cos(n*θ) or r = a * sin(n*θ)

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
a = 1
theta = np.linspace(0, 2*np.pi, 1000)

# 3 petals
ax = axes[0, 0]
n = 3
r = a * np.cos(n * theta)
x = r * np.cos(theta)
y = r * np.sin(theta)
ax.plot(x, y, 'b-', linewidth=2)
ax.fill(x, y, alpha=0.3, color='blue')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title('Rose Curve (3 petals)')

# 4 petals
ax = axes[0, 1]
n = 4
r = a * np.cos(n * theta)
x = r * np.cos(theta)
y = r * np.sin(theta)
ax.plot(x, y, 'g-', linewidth=2)
ax.fill(x, y, alpha=0.3, color='green')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title('Rose Curve (4 petals)')

# 5 petals
ax = axes[1, 0]
n = 5
r = a * np.cos(n * theta)
x = r * np.cos(theta)
y = r * np.sin(theta)
ax.plot(x, y, 'r-', linewidth=2)
ax.fill(x, y, alpha=0.3, color='red')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title('Rose Curve (5 petals)')

# 8 petals
ax = axes[1, 1]
n = 8
r = a * np.cos(n * theta)
x = r * np.cos(theta)
y = r * np.sin(theta)
ax.plot(x, y, 'm-', linewidth=2)
ax.fill(x, y, alpha=0.3, color='magenta')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title('Rose Curve (8 petals)')

plt.tight_layout()
plt.show()
```

## 7. Lissajous Curve (李萨如曲线)
**Parametric equations:**
- x = sin(a*t + δ)
- y = sin(b*t)

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
t = np.linspace(0, 2*np.pi, 1000)

# 3:2
ax = axes[0, 0]
x = np.sin(3*t)
y = np.sin(2*t)
ax.plot(x, y, 'b-', linewidth=2)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title('Lissajous Curve (3:2)')

# 5:4
ax = axes[0, 1]
x = np.sin(5*t)
y = np.sin(4*t)
ax.plot(x, y, 'g-', linewidth=2)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title('Lissajous Curve (5:4)')

# 5:3
ax = axes[1, 0]
x = np.sin(5*t)
y = np.sin(3*t)
ax.plot(x, y, 'r-', linewidth=2)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title('Lissajous Curve (5:3)')

# 7:5
ax = axes[1, 1]
x = np.sin(7*t)
y = np.sin(5*t)
ax.plot(x, y, 'm-', linewidth=2)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title('Lissajous Curve (7:5)')

plt.tight_layout()
plt.show()
```

## 8. Cycloid (摆线)
**Parametric equations:**
- x = r * (t - sin(t))
- y = r * (1 - cos(t))

```python
fig, ax = plt.subplots(figsize=(12, 6))
t = np.linspace(0, 6*np.pi, 1000)
r = 1
x = r * (t - np.sin(t))
y = r * (1 - np.cos(t))
ax.plot(x, y, 'b-', linewidth=2, label='Cycloid')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Cycloid (摆线)')
plt.show()
```

## 9. Epicycloid (外摆线)
**Parametric equations:**
- x = (R + r) * cos(t) - r * cos((R + r)*t / r)
- y = (R + r) * sin(t) - r * sin((R + r)*t / r)

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# R=5, r=1 (6 cusps)
ax = axes[0]
R, r = 5, 1
t = np.linspace(0, 2*np.pi, 1000)
x = (R + r) * np.cos(t) - r * np.cos((R + r) * t / r)
y = (R + r) * np.sin(t) - r * np.sin((R + r) * t / r)
ax.plot(x, y, 'b-', linewidth=2)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title(f'Epicycloid (R={R}, r={r})')

# R=4, r=1 (5 cusps)
ax = axes[1]
R, r = 4, 1
t = np.linspace(0, 2*np.pi, 1000)
x = (R + r) * np.cos(t) - r * np.cos((R + r) * t / r)
y = (R + r) * np.sin(t) - r * np.sin((R + r) * t / r)
ax.plot(x, y, 'g-', linewidth=2)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title(f'Epicycloid (R={R}, r={r})')

# R=3, r=1 (4 cusps)
ax = axes[2]
R, r = 3, 1
t = np.linspace(0, 2*np.pi, 1000)
x = (R + r) * np.cos(t) - r * np.cos((R + r) * t / r)
y = (R + r) * np.sin(t) - r * np.sin((R + r) * t / r)
ax.plot(x, y, 'r-', linewidth=2)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title(f'Epicycloid (R={R}, r={r})')

plt.tight_layout()
plt.show()
```

## 10. Hypocycloid (内摆线)
**Parametric equations:**
- x = (R - r) * cos(t) + r * cos((R - r)*t / r)
- y = (R - r) * sin(t) - r * sin((R - r)*t / r)

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# R=5, r=1 (5 cusps)
ax = axes[0]
R, r = 5, 1
t = np.linspace(0, 2*np.pi, 1000)
x = (R - r) * np.cos(t) + r * np.cos((R - r) * t / r)
y = (R - r) * np.sin(t) - r * np.sin((R - r) * t / r)
ax.plot(x, y, 'b-', linewidth=2)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title(f'Hypocycloid (R={R}, r={r})')

# R=4, r=1 (4 cusps - Astroid)
ax = axes[1]
R, r = 4, 1
t = np.linspace(0, 2*np.pi, 1000)
x = (R - r) * np.cos(t) + r * np.cos((R - r) * t / r)
y = (R - r) * np.sin(t) - r * np.sin((R - r) * t / r)
ax.plot(x, y, 'g-', linewidth=2)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title(f'Hypocycloid (R={R}, r={r}) - Astroid')

# R=3, r=1 (3 cusps - Deltoid)
ax = axes[2]
R, r = 3, 1
t = np.linspace(0, 2*np.pi, 1000)
x = (R - r) * np.cos(t) + r * np.cos((R - r) * t / r)
y = (R - r) * np.sin(t) - r * np.sin((R - r) * t / r)
ax.plot(x, y, 'r-', linewidth=2)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title(f'Hypocycloid (R={R}, r={r}) - Deltoid')

plt.tight_layout()
plt.show()
```

## 11. Butterfly Curve (蝴蝶曲线)
**Parametric equations:**
- x = sin(t) * (exp(cos(t)) - 2*cos(4*t))
- y = cos(t) * (exp(cos(t)) - 2*cos(4*t))

```python
fig, ax = plt.subplots(figsize=(8, 8))
t = np.linspace(0, 12*np.pi, 2000)
x = np.sin(t) * (np.exp(np.cos(t)) - 2*np.cos(4*t))
y = np.cos(t) * (np.exp(np.cos(t)) - 2*np.cos(4*t))
ax.plot(x, y, 'purple', linewidth=1.5, label='Butterfly Curve')
ax.fill(x, y, alpha=0.3, color='purple')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Butterfly Curve (蝴蝶曲线)')
plt.show()
```

## 12. Lemniscate (伯努利双纽线)
**Polar equation:**
- r² = a² * cos(2*θ)

```python
fig, ax = plt.subplots(figsize=(8, 8))
theta = np.linspace(-np.pi/4, np.pi/4, 500)
a = 1
r = a * np.sqrt(np.cos(2*theta))
x1 = r * np.cos(theta)
y1 = r * np.sin(theta)

theta = np.linspace(3*np.pi/4, 5*np.pi/4, 500)
r = a * np.sqrt(np.cos(2*theta))
x2 = r * np.cos(theta)
y2 = r * np.sin(theta)

ax.plot(x1, y1, 'b-', linewidth=2, label='Lemniscate')
ax.plot(x2, y2, 'b-', linewidth=2)
ax.fill(x1, y1, alpha=0.3, color='blue')
ax.fill(x2, y2, alpha=0.3, color='blue')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Lemniscate (伯努利双纽线)')
plt.show()
```

## 13. Spiral of Theodorus (西奥多罗斯螺旋)
**Construction:** Square root spiral

```python
fig, ax = plt.subplots(figsize=(8, 8))

x, y = 0, 0
for i in range(1, 100):
    angle = np.sqrt(i)
    x_new = x + np.cos(angle)
    y_new = y + np.sin(angle)
    ax.plot([x, x_new], [y, y_new], 'b-', linewidth=1)
    x, y = x_new, y_new

ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title('Spiral of Theodorus (西奥多罗斯螺旋)')
plt.show()
```

## 14. Fermat's Spiral (费马螺旋)
**Polar equation:**
- r = a * √θ

```python
fig, ax = plt.subplots(figsize=(8, 8))
theta = np.linspace(0, 8*np.pi, 1000)
a = 0.5
r = a * np.sqrt(theta)
x = r * np.cos(theta)
y = r * np.sin(theta)
ax.plot(x, y, 'orange', linewidth=2, label='Fermat\'s Spiral')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Fermat\'s Spiral (费马螺旋)')
plt.show()
```

## 15. Clothoid (衣摆线/欧拉螺旋)
**Parametric equations (Fresnel integrals):**
- x = ∫ cos(π*t²/2) dt
- y = ∫ sin(π*t²/2) dt

```python
from scipy.special import fresnel

fig, ax = plt.subplots(figsize=(8, 8))
t = np.linspace(-3, 3, 1000)
x, y = fresnel(t)
ax.plot(x, y, 'g-', linewidth=2, label='Clothoid')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Clothoid (衣摆线)')
plt.show()
```

## 16. Tractrix (追踪曲线)
**Parametric equations:**
- x = a * (t - tanh(t))
- y = a * sech(t)

```python
fig, ax = plt.subplots(figsize=(10, 6))
t = np.linspace(-3, 3, 1000)
a = 1
x = a * (t - np.tanh(t))
y = a / np.cosh(t)
ax.plot(x, y, 'r-', linewidth=2, label='Tractrix')
ax.fill_between(x, 0, y, alpha=0.3, color='red')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Tractrix (追踪曲线)')
plt.show()
```

## 17. Catenary (悬链线)
**Equation:**
- y = a * cosh(x/a)

```python
fig, ax = plt.subplots(figsize=(10, 6))
x = np.linspace(-4, 4, 1000)
a = 1
y = a * np.cosh(x/a)
ax.plot(x, y, 'b-', linewidth=2, label='Catenary')
ax.fill_between(x, 0, y, alpha=0.2, color='blue')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Catenary (悬链线)')
ax.set_ylim([0, 6])
plt.show()
```

## 18. Witch of Agnesi (阿涅西的女巫)
**Parametric equations:**
- x = 2*a*t
- y = 2*a / (1 + t²)

```python
fig, ax = plt.subplots(figsize=(10, 6))
t = np.linspace(-5, 5, 1000)
a = 1
x = 2*a*t
y = 2*a / (1 + t**2)
ax.plot(x, y, 'purple', linewidth=2, label='Witch of Agnesi')
ax.fill_between(x, 0, y, alpha=0.2, color='purple')
ax.axhline(y=0, color='k', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Witch of Agnesi (阿涅西的女巫)')
plt.show()
```

## 19. Cissoid (蔓叶线)
**Polar equation:**
- r = 2*a*sin(θ)*tan(θ)

```python
fig, ax = plt.subplots(figsize=(8, 8))
theta = np.linspace(-np.pi/2 + 0.1, np.pi/2 - 0.1, 1000)
a = 1
r = 2*a*np.sin(theta)*np.tan(theta)
x = r * np.cos(theta)
y = r * np.sin(theta)
ax.plot(x, y, 'cyan', linewidth=2, label='Cissoid')
ax.axvline(x=2*a, color='gray', linewidth=1, linestyle='--', label='Asymptote')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Cissoid (蔓叶线)')
ax.set_xlim([-2, 4])
ax.set_ylim([-4, 4])
plt.show()
```

## 20. Conchoid (贝壳线)
**Polar equation:**
- r = a/cos(θ) + b

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Conchoid of Nicomedes
ax = axes[0]
theta = np.concatenate([
    np.linspace(-np.pi/2 + 0.1, np.pi/2 - 0.1, 500),
    np.linspace(np.pi/2 + 0.1, 3*np.pi/2 - 0.1, 500)
])
a, b = 1, 0.5
r = a/np.cos(theta) + b
x = r * np.cos(theta)
y = r * np.sin(theta)
ax.plot(x, y, 'b-', linewidth=2, label='Conchoid (a=1, b=0.5)')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Conchoid of Nicomedes (贝壳线)')
ax.set_xlim([-3, 5])
ax.set_ylim([-3, 3])

# Different parameters
ax = axes[1]
theta = np.concatenate([
    np.linspace(-np.pi/2 + 0.1, np.pi/2 - 0.1, 500),
    np.linspace(np.pi/2 + 0.1, 3*np.pi/2 - 0.1, 500)
])
a, b = 1, 1.5
r = a/np.cos(theta) + b
x = r * np.cos(theta)
y = r * np.sin(theta)
ax.plot(x, y, 'g-', linewidth=2, label='Conchoid (a=1, b=1.5)')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Conchoid of Nicomedes (贝壳线)')
ax.set_xlim([-3, 5])
ax.set_ylim([-4, 4])

plt.tight_layout()
plt.show()
```
