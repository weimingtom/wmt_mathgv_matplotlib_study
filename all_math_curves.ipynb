# 已知数学曲线汇总
本笔记本展示并绘制多种常见的数学曲线（参数或极坐标形式）。
包括：Circle、Ellipse、Parabola、Rectangular Hyperbola、Cardioid、Limacon、Lemniscate、Cycloid、Epicycloid/Hypocycloid、Astroid 等。

使用 MathGV 和 matplotlib 进行绘制（参考 Jupyter Notebook 教程）。

```python
import numpy as np
import matplotlib.pyplot as plt

def calc_and_plot():
    curves = []
    t = np.linspace(0, 2*np.pi, 1000)

    # Circle
    curves.append(('Circle', t, lambda tt: (np.cos(tt), np.sin(tt))))

    # Ellipse
    curves.append(('Ellipse', t, lambda tt: (2*np.cos(tt), np.sin(tt))))

    # Parabola y = x^2 (parametric)
    t_p = np.linspace(-2, 2, 1000)
    curves.append(('Parabola y=x^2', t_p, lambda tt: (tt, tt**2)))

    # Rectangular hyperbola y = 1/x (avoid singularity)
    t_h = np.concatenate([np.linspace(-5, -0.2, 400), np.linspace(0.2, 5, 400)])
    curves.append(('Rect. Hyperbola y=1/x', t_h, lambda tt: (tt, 1.0/tt)))

    # Cardioid r = 1 - cos(t) (polar -> cartesian)
    curves.append(('Cardioid', t, lambda tt: ((1-np.cos(tt))*np.cos(tt), (1-np.cos(tt))*np.sin(tt))))

    # Limacon r = 1 + 0.5 cos(t)
    curves.append(('Limacon', t, lambda tt: ((1+0.5*np.cos(tt))*np.cos(tt), (1+0.5*np.cos(tt))*np.sin(tt))))

    # Lemniscate of Bernoulli (r^2 = cos(2t)) -- mask negative regions
    def lemniscate(tt):
        r2 = np.cos(2*tt)
        r = np.sqrt(np.clip(r2, 0, None))
        x = r * np.cos(tt)
        y = r * np.sin(tt)
        return x, y
    curves.append(('Lemniscate (Bernoulli)', t, lemniscate))

    # Archimedean spiral r = a + b t
    a = 0; b = 0.15
    curves.append(('Archimedean spiral', np.linspace(0, 20, 1000), lambda tt: ((a+b*tt)*np.cos(tt), (a+b*tt)*np.sin(tt))))

    # Logarithmic spiral r = exp(b t)
    curves.append(('Logarithmic spiral', np.linspace(0, 8, 1000), lambda tt: (np.exp(0.15*tt)*np.cos(tt), np.exp(0.15*tt)*np.sin(tt))))

    # Fermat's spiral r = sqrt(t)
    curves.append(("Fermat's spiral", np.linspace(0, 50, 1000), lambda tt: (np.sqrt(tt)*np.cos(tt), np.sqrt(tt)*np.sin(tt))))

    # Cycloid (rolling circle)
    t_c = np.linspace(0, 4*np.pi, 1000)
    curves.append(('Cycloid', t_c, lambda tt: (tt - np.sin(tt), 1 - np.cos(tt))))

    # Epicycloid (R outer, r rolling)
    R = 5; r = 3
    curves.append((f'Epicycloid (R={R}, r={r})', t, lambda tt: ((R+r)*np.cos(tt) - r*np.cos((R+r)/r*tt), (R+r)*np.sin(tt) - r*np.sin((R+r)/r*tt))))

    # Hypocycloid
    R = 5; r = 3
    curves.append((f'Hypocycloid (R={R}, r={r})', t, lambda tt: ((R-r)*np.cos(tt) + r*np.cos((R-r)/r*tt), (R-r)*np.sin(tt) - r*np.sin((R-r)/r*tt))))

    # Astroid x = cos^3 t, y = sin^3 t
    curves.append(('Astroid', t, lambda tt: (np.cos(tt)**3, np.sin(tt)**3)))

    # Prepare plotting grid
    n = len(curves)
    cols = 3
    rows = int(np.ceil(n / cols))
    fig, axes = plt.subplots(rows, cols, figsize=(5*cols, 4*rows))
    if rows == 1 and cols == 1:
        axes = np.array([axes])
    axes = axes.flatten()

    for ax, (name, tt, func) in zip(axes, curves):
        x, y = func(tt)
        ax.plot(x, y)
        ax.set_title(name)
        ax.set_aspect('equal', adjustable='box')
        ax.grid(True)

    # remove unused axes
    for ax in axes[n:]:
        fig.delaxes(ax)

    fig.tight_layout()
    plt.show()

if __name__ == '__main__':
    calc_and_plot()
```
