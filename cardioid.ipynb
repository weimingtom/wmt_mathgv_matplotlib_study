```python
import numpy as np
import matplotlib.pyplot as plt

# 
# 2*1*(1-cos(Theta))
Theta = np.r_[0 : 2 * np.pi : 2 * np.pi / 10000 ]
R = 2 * 1 * (1 - np.cos(Theta))

plt.polar(Theta,R)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
# 1-sin(Theta)
Theta = np.r_[0 : 2 * np.pi : 2 * np.pi / 10000 ]
R = 1 * (1 - np.sin(Theta))

plt.polar(Theta,R,linewidth=3,color='r')
plt.fill(Theta,R,color='r')
```
